```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
class binomial:
    def __init__(self, N, p): 
        self.N = N 
        self.p = p 
        
    def sample(self,n=1):
        output = []
        for rep in range(n):
            X = 0
            for i in range(self.N):
                u = random.random() 
                if u < self.p: 
                    X = X+1
            output.append(X)
        return output 
```

```python
binom = binomial(N=20,p=0.5)
out = binom.sample(100000)
plt.hist(out, bins = np.arange(-0.5,21,0.5))
```
