```python
import random
import matplotlib.pyplot as plt
```

```python
steps = 100
x = [i for i in range(steps)]
for _ in range(10):
    y = []
    start = 0

    for _ in range(steps):
        y.append(start)

        r = random.random()
        if r >= 0.5:
            start += 1
        else:
            start -= 1

    plt.plot(x,y)
    
plt.show()
```
