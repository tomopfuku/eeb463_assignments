- this short notebook provides an example of how to create a class in python and instantiate it

```python
class cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def meow(self, word = "meow"):
        print(word)
 
```

```python
our_cat = cat("Frank", 2)
```

```python
our_cat.age += 1
```

```python
our_cat.meow()
```
